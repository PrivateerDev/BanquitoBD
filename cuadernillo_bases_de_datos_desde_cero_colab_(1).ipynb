# 1. ¿Qué es una base de datos?

Una **base de datos** es un conjunto de datos organizados de forma que puedan almacenarse, consultarse y modificarse fácilmente.

Piensa en una libreta donde guardas nombres, correos y teléfonos.  
Eso ya es una forma sencilla de almacenar datos.  
La diferencia es que una base de datos permite hacerlo de manera mucho más ordenada, rápida y segura.

## Ejemplo cotidiano

Imagina un sistema escolar.  
Se necesita guardar:

- alumnos,
- profesores,
- materias,
- calificaciones,
- horarios.

Si todo eso se escribiera en archivos sueltos o en hojas dispersas, sería un caos.  
La base de datos ayuda a que toda esa información quede organizada y relacionada.

## ¿Por qué son importantes?

Porque permiten:

- almacenar grandes cantidades de información,
- evitar duplicidad innecesaria,
- buscar datos rápidamente,
- actualizar información sin rehacer todo,
- compartir información entre usuarios y sistemas.

# 2. Conceptos básicos

## Tabla
Una **tabla** almacena datos de un mismo tipo.

Ejemplo: una tabla llamada `usuarios`.

## Campo o columna
Es cada atributo que se guarda.

Ejemplo en `usuarios`:

- id
- nombre
- correo
- edad

## Registro o fila
Es cada elemento almacenado dentro de la tabla.

Ejemplo:

| id | nombre | correo | edad |
|---|---|---|---|
| 1 | Ana | ana@gmail.com | 22 |

## Clave primaria
Es un campo que identifica de manera única a cada registro.

Por ejemplo: `id`.

No debería repetirse ni quedar vacío.

## Relación
Es la conexión entre tablas.

Ejemplo:
- una tabla `clientes`
- una tabla `pedidos`

Cada pedido pertenece a un cliente.  
Ahí ya existe una relación.

---

## Ejemplo visual rápido

|---|---|---|

| 2 | María | maria@gmail.com |

Aquí:
- `id`, `nombre`, `correo` son columnas
- cada renglón es un registro
- `id` puede funcionar como clave primaria

```python
import sqlite3
import pandas as pd

conexion = sqlite3.connect("ejemplo_bd.db")
cursor = conexion.cursor()

print("Conexión realizada correctamente.")
```

# 5. Creación de una tabla

Vamos a crear una tabla llamada `usuarios`.

Esta tabla tendrá:
- `id`: identificador único
- `nombre`: nombre del usuario
- `correo`: correo electrónico
- `edad`: edad del usuario

Observa que `id` será la **clave primaria**.

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS usuarios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT NOT NULL,
    correo TEXT NOT NULL,
    edad INTEGER
)
''')

conexion.commit()
print("Tabla 'usuarios' creada correctamente.")
```

## Explicación del comando anterior

- `CREATE TABLE` crea una tabla.
- `IF NOT EXISTS` evita error si la tabla ya existe.
- `id INTEGER PRIMARY KEY AUTOINCREMENT` crea un identificador automático.
- `nombre TEXT NOT NULL` indica que el nombre es texto y no puede quedar vacío.
- `correo TEXT NOT NULL` hace lo mismo para el correo.
- `edad INTEGER` guarda números enteros.

Esto ya se parece al inicio de un sistema real.

# 6. Insertar datos: ALTAS

Cuando en un sistema registramos un nuevo usuario, producto, cliente o alumno, estamos realizando un **alta**.

En SQL esto suele hacerse con `INSERT INTO`.

```python
usuarios_iniciales = [
    ("Ana", "ana@gmail.com", 22),
    ("Luis", "luis@gmail.com", 25),
    ("María", "maria@gmail.com", 20),
    ("Carlos", "carlos@gmail.com", 28)
]

cursor.executemany(
    "INSERT INTO usuarios (nombre, correo, edad) VALUES (?, ?, ?)",
    usuarios_iniciales
)

conexion.commit()
print("Datos insertados correctamente.")
```

## ¿Qué pasó aquí?

- `INSERT INTO usuarios` indica en qué tabla guardaremos los datos.
- `(nombre, correo, edad)` indica qué columnas llenaremos.
- `VALUES (?, ?, ?)` usa parámetros para insertar valores de forma ordenada.
- `executemany()` permite insertar varios registros a la vez.

En un sistema web, esta operación puede ocurrir cuando el usuario llena un formulario y presiona **Guardar**.

# 7. Consultar datos: BÚSQUEDAS

La búsqueda de información se realiza con `SELECT`.

Primero veremos todos los registros de la tabla.

```python
consulta = "SELECT * FROM usuarios"
df_usuarios = pd.read_sql_query(consulta, conexion)
df_usuarios
```

## Explicación

- `SELECT` significa seleccionar.
- `*` significa “todas las columnas”.
- `FROM usuarios` indica de qué tabla se obtendrán los datos.

Esta es la operación más común en pantallas de consulta, listados y reportes.

# 8. Búsqueda con condiciones

Ahora consultaremos solo usuarios mayores de 22 años.

```python
consulta = "SELECT * FROM usuarios WHERE edad > 22"
df_filtrado = pd.read_sql_query(consulta, conexion)
df_filtrado
```

## Explicación

La cláusula `WHERE` se usa para aplicar condiciones.

Ejemplos:
- `WHERE edad > 22`
- `WHERE nombre = 'Ana'`
- `WHERE correo LIKE '%gmail.com%'`

Esto es muy útil cuando el sistema permite búsquedas por nombre, edad, correo, categoría, fecha, etc.

# 9. Actualizar datos: CAMBIOS

Cuando modificamos información existente, usamos `UPDATE`.

Supongamos que Ana cambió de edad o capturamos mal el dato y queremos corregirlo.

```python
cursor.execute('''
UPDATE usuarios
SET edad = 23
WHERE nombre = 'Ana'
''')

conexion.commit()
print("Registro actualizado correctamente.")
```

```python
pd.read_sql_query("SELECT * FROM usuarios", conexion)
```

## Explicación

- `UPDATE usuarios` indica qué tabla se modificará.
- `SET edad = 23` define el nuevo valor.
- `WHERE nombre = 'Ana'` indica qué registro se modificará.

**Advertencia importante:**  
Si haces `UPDATE` sin `WHERE`, podrías modificar **todos** los registros.  
Eso ya no sería una práctica de bases de datos. Sería deporte extremo.

# 10. Eliminar datos: BAJAS

Cuando quitamos un registro, usamos `DELETE`.

Eliminaremos a Carlos de la tabla.

```python
cursor.execute('''
DELETE FROM usuarios
WHERE nombre = 'Carlos'
''')

conexion.commit()
print("Registro eliminado correctamente.")
```

```python
pd.read_sql_query("SELECT * FROM usuarios", conexion)
```

## Explicación

- `DELETE FROM usuarios` indica de qué tabla se eliminará.
- `WHERE nombre = 'Carlos'` define qué registro borrar.

**Advertencia:**  
Igual que con `UPDATE`, si haces `DELETE` sin `WHERE`, podrías eliminar todos los registros.

# 12. Diseño básico de una base de datos para un sistema web

Vamos a imaginar que el proyecto será un sistema web para gestionar **productos**.

Cada producto tendrá:

- id
- nombre
- categoría
- precio
- stock

Primero creamos una nueva tabla.

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS productos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT NOT NULL,
    categoria TEXT NOT NULL,
    precio REAL NOT NULL,
    stock INTEGER NOT NULL
)
''')

conexion.commit()
print("Tabla 'productos' creada correctamente.")
```

## ¿Por qué esta tabla funciona bien?

Porque tiene estructura clara:

- un identificador único,
- datos descriptivos,
- datos numéricos,
- y campos obligatorios.

Ese tipo de diseño ayuda a que el sistema web sea ordenado desde el inicio.

# 13. Probar CRUD en la tabla productos
## Altas

```python
productos = [
    ("Laptop", "Tecnología", 14500.50, 10),
    ("Mouse", "Tecnología", 250.00, 35),
    ("Teclado", "Tecnología", 480.00, 20),
    ("Silla", "Mobiliario", 1800.00, 8)
]

cursor.executemany(
    "INSERT INTO productos (nombre, categoria, precio, stock) VALUES (?, ?, ?, ?)",
    productos
)

conexion.commit()
pd.read_sql_query("SELECT * FROM productos", conexion)
```

## Búsqueda

```python
pd.read_sql_query(
    "SELECT * FROM productos WHERE categoria = 'Tecnología'",
    conexion
)
```

## Cambios

```python
cursor.execute('''
UPDATE productos
SET stock = 15
WHERE nombre = 'Laptop'
''')
conexion.commit()

pd.read_sql_query("SELECT * FROM productos", conexion)
```

## Bajas

```python
cursor.execute('''
DELETE FROM productos
WHERE nombre = 'Mouse'
''')
conexion.commit()

pd.read_sql_query("SELECT * FROM productos", conexion)
```

# 15. ¿Cómo se conecta esto con una página web?

En un sistema web normalmente existen tres partes:

## 1. Frontend
Es la parte visual que ve el usuario:
- formularios

Puede hacerse con:

# 16. Ejemplo simple de lógica CRUD en Python

Aquí mostraremos una función por cada operación básica.  
No es aún una aplicación web completa, pero sí permite entender la lógica que usaría el backend.

```python
def alta_producto(nombre, categoria, precio, stock):
    cursor.execute(
        "INSERT INTO productos (nombre, categoria, precio, stock) VALUES (?, ?, ?, ?)",
        (nombre, categoria, precio, stock)
    )
    conexion.commit()

def buscar_productos():
    return pd.read_sql_query("SELECT * FROM productos", conexion)

def actualizar_precio(nombre, nuevo_precio):
    cursor.execute(
        "UPDATE productos SET precio = ? WHERE nombre = ?",
        (nuevo_precio, nombre)
    )
    conexion.commit()

def eliminar_producto(nombre):
    cursor.execute(
        "DELETE FROM productos WHERE nombre = ?",
        (nombre,)
    )
    conexion.commit()

alta_producto("Monitor", "Tecnología", 3200.00, 7)
buscar_productos()
```

## ¿Qué representa esto?

Representa la lógica que más adelante puede conectarse con:

- un botón de guardar,
- una caja de búsqueda,
- una pantalla de edición,
- un botón de eliminar.

Es decir, aquí ya está la idea central del backend.
